```python
import re
```

```python
def convert(i):
    if type(i) != str: 
        return {}
    return {x.split(':')[0]:x.split(':')[1] for x in i.split('\n') if x}

```

```python
base['new_info'] = base['info'].apply(convert)
```

```python
base = pd.concat([base, pd.json_normalize(base['new_info'])], axis=1)
```
