## <b>Tratando/transformando algumas colunas</b>
<hr>

```python
lista_atores = base['actor_name'].apply(eval) # pega todo valor linha a linha e transforma em um elemento do python, no caso vira uma lista
```

```python
base['actor_count'] = [len(x) for x in lista_atores] # criando uma coluna com a contagem dos atores
```

```python
# definindo função para 'explodir' coluna INFO
def convert(i):
    if type(i) != str: 
        return {}
    return {x.split(':')[0].strip():x.split(':')[1].strip() for x in i.split('\n') if x}
```

```python
base['new_info'] = base['info'].apply(convert)
base['new_info'][0]
```

```python
base = pd.concat([base, pd.json_normalize(base['new_info'])], axis=1)
base.columns
```

```python
base['Box Office (Gross USA)'].value_counts()
```

```python
# definindo função que transforma coluna do Box Office em float64
def transform_Box_Office(i):
    if str(i) == 'nan':
        return np.nan
    elif 'M' in str(i):
        return eval(str(i).replace('$','').replace('M','* 1000000'))
    elif 'K' in str(i):
        return eval(str(i).replace('$','').replace('K','* 1000'))
```

```python
base['Box Office (Gross USA)'] = base['Box Office (Gross USA)'].apply(transform_Box_Office)
```

```python
#verificando se há algum caracter diferente de float64 na coluna Box Office (por conta dos cifrões diferentes)
base[base['Box Office (Gross USA)'].astype(str).str.isalpha()]['Box Office (Gross USA)'].dropna()
```

```python
# transformando a coluna Release Date (Streaming) em datetime64
base['Release Date (Streaming)'] = pd.to_datetime(base['Release Date (Streaming)'].str.replace(',','').str.replace(' ', '-'), format='%b-%d-%Y')
base['Release Date (Streaming)']
```

```python
# transformando a coluna Release Date (Streaming) em datetime64
# Usando saídas no replace \s (espaços) \w palavras $ (no fim da string)
base['Release Date (Theaters)'] = pd.to_datetime(base['Release Date (Theaters)'].str.replace(r'\s+\w+$', '', regex=True).str.replace(',','').str.replace(' ','-'), format='%b-%d-%Y')
base['Release Date (Theaters)']
```
