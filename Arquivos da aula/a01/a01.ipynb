# Exercício

Faça uma função para gerar os $n$ ($n\ge3$) primeiros elementos da sequência de Fibonacci, supondo que o primeiro e o segundo elemento seja 0 e 1, respectivamente.

```python
def fibonacci(n):
    if n < 3:
        print("Preencha um valor maior que 2!")
    else:
        seq = [0,1]
        for i in range(1,n-1):
            seq.append(seq[i] + seq[i-1])
        print(seq)
        return seq[n-1]

print(fibonacci(20))
```

## Solução

![picture](https://drive.google.com/thumbnail?id=18vcvd-5xrWh_J0sq8dbrryJYhv1SxtCU&sz=w1200)

```python
def fibonacci(a, b, n):
  lista = [a, b]
  n = n - 2
  
  for i in range(n):
    c = a + b
    lista.append(c)
    
    a = b
    b = c

  return lista

n = 15
a = 0
b = 1

print(fibonacci(a, b, n))
```
