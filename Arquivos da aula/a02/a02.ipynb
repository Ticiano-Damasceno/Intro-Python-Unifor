```python
with open('teste.txt', 'r') as arquivo:  #O Parâmetro 'r' para usar a função open na forma de read  
    for i, linha in enumerate(arquivo):
        campo = linha.split()
        print(i, campo[0], campo[1])
```

## 2.2 Séries

![picture](https://drive.google.com/thumbnail?id=1krgr7CPvTyzA_GBciqvocWZWx-A8h3f7&sz=w300)

```python
import pandas as pd
import numpy as np

apples = [3, 2, 0, 1]
s = pd.Series(apples,
              name = 'apples',
              dtype = np.int64)

print(s)
# print(type(s))
# print(dir(s))

print(s.head(2)) #primeiros
print(s.tail(2)) #últimos

print(s.ndim) #nº de dimensões
print(s.shape)
print(s.mean())
print(s.max())
print(s.min())
print(s[2])
```

## 2.3 DataFrames (Planilhas)

## 2.6 Iterando sobre DataFrames

```python
d = {'apples': {'A': 3, 'B': 2, 'C': 0, 'D': 1},
     'oranges': {'A': 0, 'B': 3, 'C': 7, 'D': 2},
     'bananas': {'A': 1, 'B': 3, 'C': 5, 'D': 4},
     'avocados': {'A': 9, 'B': 0, 'C': 0, 'D': 1}}

df = pd.DataFrame(d)
print(df)
```

```python
for indice in df.index:
    print(indice, df.loc[indice, 'apples'])

# for indice in df.index:
#     print(indice, df['apples'][indice])

# for i in range(df.shape[0]):
#     print(i, df.iloc[i, 0])

for indice, linha in df.iterrows():
    print(indice, linha['apples'])
```

## 2.8 Relacionando DataFrames

![picture](https://drive.google.com/thumbnail?id=1YRUghTjQRDAoXXpKBKsSrINUJ4o_v0Z0&sz=w1200)

```python
d = {'Customer_id': pd.Series([1, 2, 3, 4, 5, 7]),
     'Product': pd.Series(['Oven', 'Oven', 'Oven', 'Television', 'Television', 'Television'])}
df0 = pd.DataFrame(d)
print(df0)

d = {'Customer_id': pd.Series([2, 4, 6]),
     'State': pd.Series(['California', 'California', 'Texas'])}
df1 = pd.DataFrame(d)
print(df1)
```

```python
print(pd.merge(df0, df1, on = 'Customer_id', how = 'inner')) #inner join

#print(pd.merge(df0, df1, on = 'Customer_id', how = 'outer')) #outter join
# print(pd.merge(df0, df1, on = 'Customer_id', how = 'left')) # left join
# print(pd.merge(df0, df1, on = 'Customer_id', how = 'right')) # right join
```
